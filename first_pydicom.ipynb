```python
from pydicom import dcmread
from pydicom.data import get_testdata_file
```

```python
path = get_testdata_file('CT_small.dcm')
ds = dcmread(path)

type(ds.PixelData)
```

```python
import matplotlib.pyplot as plt
arr = ds.pixel_array

plt.figure(figsize = (10, 10))
plt.axis('off')
plt.imshow(arr, cmap='gray')
plt.show()
```
